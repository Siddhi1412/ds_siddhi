### Final Observations

1. When the market is in **Fear** or **Extreme Fear**
   - average trade size reduces
   - more SELL positions are taken
   - traders try to reduce exposure

2. When the market is in **Greed** or **Extreme Greed**
   - BUY positions increase
   - traders take larger size
   - average closed PnL is better

3. Overall — trader behaviour follows market sentiment very clearly.

### Final Conclusion
Market emotions are not just visible in charts — they reflect directly in how traders place their orders, take risks and close positions.

This means sentiment signals can be used as leading indicators to adjust position sizing + risk.
